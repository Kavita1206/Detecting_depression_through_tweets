# Task 2


Dropping 'ItemID' & 'SentimentSource'

```python
df = df.drop(['ItemID','SentimentSource'],axis=1)
```

```python
df.head()
```

Removing Punctuations

```python
import string
string.punctuation
```

```python
punctuation_list = list(string.punctuation)
```

```python
def remove_punctuations(text,punc_list=punctuation_list):
  for punc in punc_list:
    if punc in text:
      text = text.replace(punc,' ')
  return text.strip()
```

```python
df['SentimentText'] = df['SentimentText'].apply(remove_punctuations)
df.head()
```

Removing URLs

```python
 df['SentimentText'] = df['SentimentText'].str.replace('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', ' ')
```

```python
df.head()
```

Lower Casing

```python
df['SentimentText'] = df['SentimentText'].str.lower()
```

```python
df.head()
```

Tokenization

```python
def tokenization(text):
  return text.split()
```

```python
df['msg_tokenized'] = df['SentimentText'].apply(tokenization)
```

```python
df.head()
```

```python
import nltk
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
```

```python
df['After removing stopwords']=df['msg_tokenized'].apply(lambda x: [item for item in x if item not in stopwords])
```

```python
df.head()
```

Stemming


```python
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()
```

```python
def stemming(text):
  stem_text = [porter_stemmer.stem(word) for word in text]
  return stem_text
```

```python
df['msg_stemmed'] = df['After removing stopwords'].apply(stemming)
```

```python
df
```

Lemmatization

```python
from nltk.stem import WordNetLemmatizer
nltk.download('omw-1.4')
nltk.download('wordnet')
wordnet_lemmatizer = WordNetLemmatizer()
def lemmatizer(text):
  lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
  return lemm_text
```

```python
df['SentimentText_lemmatized']=df['After removing stopwords'].apply(lemmatizer)
df['msg_lemmatized'] = df['SentimentText_lemmatized'].agg(lambda x: ','.join(map(str,x))).str.replace(',',' ')
```

```python
df.head()
```
